## 2. Entrainer un Tokenizer - ici Byte Pair Encoding (BPE)

Avant d’entraîner un tokenizer, plusieurs étapes sont importantes dans le traitement du texte. Voici un aperçu concis des concepts généraux utilisés dans le code ci-dessous:

---

### **1️⃣ Normalisation (non utilisée ici)**
La **normalisation** consiste à nettoyer et harmoniser le texte avant de le découper en sous-mots si nécessaire. Elle peut inclure :  
- ✅ Suppression des espaces inutiles  
- ✅ Mise en minuscule  
- ✅ Suppression des accents  
- ✅ Normalisation Unicode (ex. **NFC** : transforme les caractères décomposés en une seule lettre, ex. `"e´"` → `"é"` ; **NFKC** : en plus, remplace certains symboles par leurs équivalents standards, ex. `"①"` → `"1"`).

📌 **Dans notre code, cette étape n’est pas utilisée**, mais elle peut être essentielle dans d'autres modèles.

---

### **2️⃣ Pré-tokenisation**  
La **pré-tokenisation** consiste à découper le texte en unités de base avant d'appliquer une tokenisation comme BPE.  
Dans le code, on utilise une pré-tokenisation basée sur les espaces, ce qui signifie que **le texte est d’abord segmenté par les espaces** avant d’être traité par l’algorithme BPE.  

#### **Exemples de pré-tokenisation possibles** :
- **Basée sur les espaces** (`Whitespace`) → Segmente uniquement aux espaces.  
- **Basée sur les espaces avec suppression de la ponctuation** (`WhitespaceSplit`) → Supprime la ponctuation.  
- **Basée sur les sous-mots et les espaces** (`ByteLevel`) → Préserve les espaces sous forme de symboles spéciaux (ex. `"Hello world"` → `["ĠHello", "Ġworld"]`).  
- **Basée sur la segmentation BERT** (`BertPreTokenizer`) → Sépare les mots et la ponctuation (ex. `"C'est génial!"` → `["C", "'", "est", "génial", "!"]`).

---

### **3️⃣ Entrainement du Tokenizer**
(Ici on focalise sur la formation ou l'apprentissage des tokens du vocabulaire avec BPE - Pour les détails sur BPE voir les Slides de la Semaine 5 Moodle ou [Cours Hugging Face sur BPE](https://huggingface.co/learn/nlp-course/en/chapter6/5)).

Le **BPE** est un algorithme de tokenization basé sur la fusion des sous-mots.  

#### **📌 Étapes d'entraînement du BPE :**
1. **Départ :** On commence avec un vocabulaire de base contenant uniquement les caractères uniques du corpus.
2. **Appprentissage :**
  - On repère la paire de tokens la plus fréquente et on la fusionne en un nouveau token.
  - Répétition du processus jusqu’à atteindre la taille de vocabulaire définie.
3. **Stockage :** On enregistre les règles de fusion apprises et le vocabulaire final.

Lorsque nous entraînons un tokenizer BPE, nous devons définir deux éléments importants :

 - Une taille de vocabulaire cible (`vocab_size`)
 - Des tokens spéciaux (`<pad>`, `<s>`, `</s>`, `<unk>`)

Certains tokens ne viennent pas du texte lui-même, mais sont ajoutés pour aider le modèle à fonctionner correctement.

| **Token**  | **Utilité** |
|------------|------------|
| `<pad>`    | Utilisé pour remplir (padding) les phrases courtes afin qu'elles aient toutes la même longueur. |
| `<s>`      | Marque le début d’une phrase. |
| `</s>`     | Marque la fin d’une phrase. |
| `<unk>`    | Remplace un mot inconnu qui ne figure pas dans le vocabulaire du tokenizer. |

---

### **4️⃣ Post-processing**
Une fois la tokenization effectuée, une étape de **post-processing** permet d’ajouter des **tokens spéciaux** pour structurer la séquence, comme `<s>` et `</s>` pour encadrer le texte.  
- Ajoute `<s>` et `</s>` autour des phrases, selon un format spécifique :
  - Phrase unique : `<s>` phrase `</s>`
  - Paire de phrases : `<s>` phrase1 `</s>` phrase2 `</s>`

Cela garantit que chaque séquence respecte un format standard, ce qui est essentiel pour entraîner des modèles de type Transformer ultérieurement.
