## 2. Entrainer un Tokenizer - ici Byte Pair Encoding (BPE)

Avant d’entraîner un tokenizer, plusieurs étapes sont importantes dans le traitement du texte. Voici un aperçu concis des concepts généraux utilisés dans le code ci-dessous:

---

### **1️⃣ Normalisation (non utilisée ici)**
La **normalisation** consiste à nettoyer et harmoniser le texte avant de le découper en sous-mots si nécessaire. Elle peut inclure :  
- ✅ Suppression des espaces inutiles  
- ✅ Mise en minuscule  
- ✅ Suppression des accents  
- ✅ Normalisation Unicode (ex. **NFC** : transforme les caractères décomposés en une seule lettre, ex. `"e´"` → `"é"` ; **NFKC** : en plus, remplace certains symboles par leurs équivalents standards, ex. `"①"` → `"1"`).

📌 **Dans notre code, cette étape n’est pas utilisée**, mais elle peut être essentielle dans d'autres modèles.

---

### **2️⃣ Pré-tokenisation**  
La **pré-tokenisation** consiste à découper le texte en unités de base avant d'appliquer une tokenisation comme BPE.  
Dans le code, on utilise une pré-tokenisation basée sur les espaces, ce qui signifie que **le texte est d’abord segmenté par les espaces** avant d’être traité par l’algorithme BPE.  

#### **Exemples de pré-tokenisation possibles** :
- **Basée sur les espaces** (`Whitespace`) → Segmente uniquement aux espaces.  
- **Basée sur les espaces avec suppression de la ponctuation** (`WhitespaceSplit`) → Supprime la ponctuation.  
- **Basée sur les sous-mots et les espaces** (`ByteLevel`) → Préserve les espaces sous forme de symboles spéciaux (ex. `"Hello world"` → `["ĠHello", "Ġworld"]`).  
- **Basée sur la segmentation BERT** (`BertPreTokenizer`) → Sépare les mots et la ponctuation (ex. `"C'est génial!"` → `["C", "'", "est", "génial", "!"]`).

---

### **3️⃣ Entrainement du Tokenizer**
(Ici on focalise sur la formation ou l'apprentissage des tokens du vocabulaire avec BPE - Pour les détails sur BPE voir les Slides de la Semaine 5 Moodle ou [Cours Hugging Face sur BPE](https://huggingface.co/learn/nlp-course/en/chapter6/5)).

Le **BPE** est un algorithme de tokenization basé sur la fusion des sous-mots.  

#### **📌 Étapes d'entraînement du BPE :**
1. **Départ :** On commence avec un vocabulaire de base contenant uniquement les caractères uniques du corpus.
2. **Appprentissage :**
  - On repère la paire de tokens la plus fréquente et on la fusionne en un nouveau token.
  - Répétition du processus jusqu’à atteindre la taille de vocabulaire définie.
3. **Stockage :** On enregistre les règles de fusion apprises et le vocabulaire final.

Lorsque nous entraînons un tokenizer BPE, nous devons définir deux éléments importants :

 - Une taille de vocabulaire cible (`vocab_size`)
 - Des tokens spéciaux (`<pad>`, `<s>`, `</s>`, `<unk>`)

Certains tokens ne viennent pas du texte lui-même, mais sont ajoutés pour aider le modèle à fonctionner correctement.

| **Token**  | **Utilité** |
|------------|------------|
| `<pad>`    | Utilisé pour remplir (padding) les phrases courtes afin qu'elles aient toutes la même longueur. |
| `<s>`      | Marque le début d’une phrase. |
| `</s>`     | Marque la fin d’une phrase. |
| `<unk>`    | Remplace un mot inconnu qui ne figure pas dans le vocabulaire du tokenizer. |

---

### **4️⃣ Post-processing**
Une fois la tokenization effectuée, une étape de **post-processing** permet d’ajouter des **tokens spéciaux** pour structurer la séquence, comme `<s>` et `</s>` pour encadrer le texte.  
- Ajoute `<s>` et `</s>` autour des phrases, selon un format spécifique :
  - Phrase unique : `<s>` phrase `</s>`
  - Paire de phrases : `<s>` phrase1 `</s>` phrase2 `</s>`

Cela garantit que chaque séquence respecte un format standard, ce qui est essentiel pour entraîner des modèles de type Transformer ultérieurement.

---

```python
# Étape 2 : Entraîner un tokenizer BPE
def train_bpe_tokenizer(texts, vocab_size=2048):
    """
    Entraîner un tokenizer basé sur l'encodage Byte Pair Encoding (BPE) sur les examples de texte fournis.

    Args :
        texts (list) : Liste de chaînes de texte servant à l'entraînement du tokenizer.
        vocab_size (int) : Taille du vocabulaire du tokenizer.

    Returns :
        Tokenizer : Un tokenizer BPE entraîné.
    """
    # Initialiser un tokenizer BPE
    tokenizer = Tokenizer(BPE())
    # Utiliser une pre-tokenisation basée sur les espaces avant d'appliquer BPE
    tokenizer.pre_tokenizer = Whitespace()

    # Définir un entraîneur avec une taille de vocabulaire spécifiée et rajout des tokens spéciaux
    trainer = BpeTrainer(vocab_size=vocab_size, special_tokens=["<pad>", "<s>", "</s>", "<unk>"])

    # TODO : Entraîner le tokenizer en utilisant les textes fournis
    tokenizer.train_from_iterator(texts, trainer)  # Remplissez avec les paramètres corrects


    # Définir le post-traitement pour les séquences de tokens

    #<s> et </s> : Marque le début et la fin des phrases.
    #$A : Première séquence.
    #$B:1 : Deuxième séquence (on lui associe un token id ou type 1).
    #</s>:1 : Fin de la deuxième séquence (token_type_id = 1).
    #Si vous utilisez ultérieurement un modèle comme BERT, T5 ou BART, ce format est utilisé pour structurer les entrées et séparer les segments dans une tâche de NLP.

    tokenizer.post_processor = TemplateProcessing(
        single="<s> $A </s>",
        pair="<s> $A </s> $B:1 </s>:1",
        special_tokens=[("<s>", 1), ("</s>", 2)]
    )

    return tokenizer
```

# Partie 2 : Préparation des données  

Dans cette section, vous allez implémenter une classe `TweetDataset` dans PyTorch afin de prétraiter et charger efficacement le jeu de données (dataset) pour l'entraînement du modèle.  

Vous allez :  

1. Construire un vocabulaire pour les modèles Bag-of-Words (BoW).  
2. Implémenter deux variantes de BoW :  
   - Représentation booléenne (indiquant la présence ou l'absence d'un mot).  
   - Représentation en fréquence des termes (comptant le nombre d'occurrences des mots).  
3. Extraire les caractéristiques du texte d'entrée et préparer le jeu de données pour l'entraînement du modèle.


```python
# Étape 1 : Définir une classe de jeu de données personnalisée
class TweetDataset(Dataset):
    def __init__(self, file_path, tokenizer, is_train=True, vocab=None):
        """
        Cette classe de jeu de données charge les tweets à partir d'un fichier, tokenise le texte et construit un vocabulaire.

        Args :
            file_path (str) : Chemin du fichier du jeu de données au format JSONL.
            tokenizer (Tokenizer) : Un tokenizer BPE entraîné.
        """
        self.data = []

        # Charger le jeu de données à partir d'un fichier JSONL
        with open(file_path, "r") as f:
            for line in f:
                self.data.append(json.loads(line))

        self.tokenizer = tokenizer

        if is_train:
            # Construire le vocabulaire à partir du jeu de données
            vocab_set = set()
            for sample in self.data:
                text = sample["text"]
                text_tokens = self.tokenizer.encode(text).tokens
                vocab_set.update(text_tokens)

            # TODO : Remplacez None et construisez un vocabulaire mappant les tokens à des indices uniques
            self.vocab = {token: i for i, token in enumerate(sorted(vocab_set))}

        else:
            self.vocab = vocab


    # Étape 2 : Implémenter la représentation Bag-of-Words booléenne
    def bow_boolean(self, tokens, vocab):
        """
        Convertit une liste de tokens en un vecteur BoW binaire.

        Args :
            tokens (list) : Liste de tokens extraits d'un tweet.
            vocab (dict) : Mappage du vocabulaire.

        Returns :
            torch.Tensor : Vecteur binaire représentant la présence des mots dans le vocabulaire.
        """
        # Initialiser un vecteur de zéros
        vector = torch.zeros(len(vocab))
        # Utiliser un ensemble pour garantir une présence binaire (0 ou 1)
        for token in set(tokens):
            if token in vocab:
                # TODO : Remplacez None et remplissez correctement le vecteur avec 1
                vector[vocab[token]] = 1
        return vector

    # Étape 3 : Implémenter la représentation Bag-of-Words (BoW) en fréquence des termes
    def bow_term_frequency(self, tokens, vocab):
        """
        Calcule la fréquence des termes (BoW basé sur le comptage des mots) pour un texte tokenisé donné.

        Args :
            tokens (list) : Liste de tokens extraits d'un tweet.
            vocab (dict) : Mappage du vocabulaire.

        Returns :
            torch.Tensor : Vecteur représentant les fréquences des termes.
        """
        vector = torch.zeros(len(vocab))
        token_counts = Counter(tokens)
        for token, count in token_counts.items():
            if token in vocab:
                # TODO : Remplacez None et remplissez correctement le vecteur avec les fréquences des termes
                vector[vocab[token]] = count
        return vector


    # Étape 4 : Obtenir la longueur du jeu de données
    def __len__(self):
        """Retourne le nombre d'échantillons dans le jeu de données."""
        return len(self.data)


    # Étape 5 : Récupérer un élément du jeu de données
    def __getitem__(self, idx):
        """
        Récupère un échantillon/exemple du jeu de données et génère son vecteur de caractéristiques.

        Args :
            idx (int) : Index de l'échantillon/exemple.

        Returns :
            dict : Dictionnaire contenant le texte tokenisé et son étiquette.
        """
        text = self.data[idx]["text"]
        label = self.data[idx]["label"]
        tokenized_text = self.tokenizer.encode(text).tokens

        # TODO : Extraire les features / caractéristiques en utilisant le BoW booléen
        features = self.bow_boolean(tokenized_text, self.vocab)


        # Conversion des caractéristiques et du label en tenseurs PyTorch
        return {
    "features": features.clone().detach(),
    "label": torch.tensor(label)
}
```

```python
# Etape 1: Definir le modèle de réseaux de neurones
class SentimentModel(nn.Module):
    def __init__(self, input_dim):
        """
        Crée un réseau de neurones feedforward pour la classification des sentiments.

        Args :
            input_dim (int) : Taille du vecteur de caractéristiques en entrée.
        """
        # Initialiser nn.Module pour que PyTorch enregistre correctement
        # les couches et paramètres du modèle
        super(SentimentModel, self).__init__()

        # TODO : Remplissez les espaces vides avec les dimensions appropriées
        self.fc1 = nn.Linear(input_dim, 1024)
        self.fc2 = nn.Linear(1024, 512)
        self.fc3 = nn.Linear(512, 128)
        self.fc4 = nn.Linear(128, 3)
        self.relu = nn.ReLU()


    def forward(self, x):
        x = self.fc1(x)
        x = self.relu(x)
        x = self.fc2(x)
        x = self.relu(x)
        x = self.fc3(x)
        x = self.relu(x)
        x = self.fc4(x)
        return x
```

**Brève introduction à l'optimisation Adam**  

**Adam (Adaptive Moment Estimation)** est un algorithme d'optimisation largement utilisé pour l'entraînement des réseaux de neurones. Il est populaire en raison de son efficacité et de sa nature adaptative, combinant des idées du SGD avec momentum et de RMSprop pour ajuster automatiquement le taux d'apprentissage.  

- Une moyenne pondérée des gradients passés pour fournir une direction (premier moment).  
- Une moyenne pondérée des gradients passés au carré pour adapter le taux d'apprentissage à chaque paramètre (second moment).  

Cela permet à Adam d'ajuster dynamiquement le taux d'apprentissage pour chaque paramètre au cours de l'entraînement.  

Kingma, D. P., & Ba, J. (2014). [Adam: A Method for Stochastic Optimization](https://arxiv.org/abs/1412.6980).  


```python
# Étape 2 : Entraîner le modèle
def train_model(model, dataloader, optimizer_type="adam", epochs=5):
    """
    Entraîne le modèle de classification des sentiments en utilisant l'entropie croisée comme fonction de perte et un optimiseur.

    Args :
        model : Modèle de réseau de neurones.
        dataloader : Chargeur de données d'entraînement.
        optimizer_type : Choix de l'optimiseur ('adam' ou 'sgd').
        epochs : Nombre d'époques d'entraînement.
    """
    criterion = nn.CrossEntropyLoss()
    optimizer = optim.Adam(model.parameters()) if optimizer_type == "adam" else optim.SGD(model.parameters(), lr=0.01)

    for epoch in range(epochs):
        for batch in dataloader:
            features_batch = batch["features"]
            label_batch = batch["label"]

            # TODO : Remplacez None et initialisez l'optimiseur avec un gradient nul
            optimizer.zero_grad()


            # TODO : Remplacez None et passez les features/caractéristiques à travers le modèle et calculez le cout
            outputs = model(features_batch)
            loss = criterion(outputs, label_batch)


            # TODO : Remplacez None et effectuez la rétropropagation ainsi que la mise à jour des poids de l'optimiseur
            loss.backward()
            optimizer.step()

        print(f"Époque : {epoch}, Perte : {loss.item()}")

    # Sauvegarder les poids entraînés du modèle (state_dict) dans un fichier pour pouvoir le recharger plus tard
    torch.save(model.state_dict(), "sentiment_model.pth")
    print("Modèle entraîné et sauvegardé !")
```

```python
# Étape 3 : Évaluer le modèle
def evaluate_model(model, dataloader):
    """
    Charger le modèle entraîné et l'évaluer sur le jeu de test.

    Args :
        model : Modèle de réseau de neurones.
        dataloader : Chargeur de données contenant les exemples de test.

    Cette fonction évalue la performance du modèle sur le jeu de test en calculant
    l'exactitude (accuracy) et le score F1.
    """

    # Charger les poids du modèle entraîné depuis le fichier sauvegardé
    model.load_state_dict(torch.load("sentiment_model.pth"))

    # Mettre le modèle en mode évaluation (désactive le dropout et la mise à jour des poids)
    model.eval()

    # Listes pour stocker les étiquettes réelles et les prédictions du modèle
    y_true, y_pred = [], []

    # Désactiver le calcul du gradient en phase d'évaluation/test pour économiser de la mémoire et accélérer l'évaluation
    with torch.no_grad():
        for batch in dataloader:
            # Extraire les caractéristiques et les étiquettes de la batch
            features_batch = batch["features"]
            label_batch = batch["label"]

            # Passer les données à travers le modèle pour obtenir les prédictions
            outputs = model(features_batch)

            # Convertir les sorties du modèle en classes prédites (classe avec la plus haute probabilité)
            predictions = torch.argmax(outputs, dim=1).tolist()

            # TODO : Remplacez None pour stocker correctement les étiquettes réelles et les prédictions
            y_true.extend(label_batch.tolist())  # Ajouter les vraies étiquettes
            y_pred.extend(predictions)  # Ajouter les prédictions du modèle

    # Calculer la précision (accuracy) en pourcentage
    accuracy = accuracy_score(y_true, y_pred) * 100

    # Calculer le score F1 macro (moyenne des scores F1 pour chaque classe)
    f1 = f1_score(y_true, y_pred, average="macro") * 100

    # Afficher les métriques d'évaluation
    print(f"Précision : {accuracy:.2f}%, Score F1 : {f1:.2f}%")
```
